# **Árvore de decisão para análise de crédito**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

# **importando os dados da análise**

```python
data = pd.read_csv('/content/drive/MyDrive/credito_correto.csv')
```

# **Os dados contém 2000 registros para análise e 4 colunas (rendimento,idade,dívida e inadimplencia)**

```python
data
```

# **rendimento, idade e dívida serão usados para a máquina classificar como possiveis inadimplentes**

```python
x = data[['idade','dívida','rendimento']]
y = data['inadimplencia']
```

# **usaremos 500 registros para treinar nossa maquina e 1500 registros para testar a acertividade da máquina**

```python
x_treino = x[:500].values
y_treino = y[:500].values
x_teste = x[500:].values
y_teste = y[500:].values
```

```python
modelo = DecisionTreeClassifier() 
modelo.fit(x_treino, y_treino)
previsoes = modelo.predict(x_teste)

acuracia = accuracy_score(y_teste, previsoes) * 100
print(f'a acertividade foi de {acuracia}')
```

# **nesta análise obtivemos uma acertividade de 96,2% com nossa Árvore de decisão** (por:Lucas Mareglia)
